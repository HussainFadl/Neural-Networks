Combining the traditional (Linear regression modeling) to (Neural networks) produces (Logistic regression)...  

Let's recall formula ...  
$$a + C_1 * X_1 + {....} + C_p * X_p $$

Where $C$ is the input data.  
      $X$ is the coeffecient.    
      $a$ is the intercept parameter

However, the difference between linear regression and logistic regression lies in what happens after the above linear function has been calculated.  
In linear regression, we're done and the linear function is the output.  
In Logistic regression,  we take the result of the linear function and apply a special non-linear function called the $(sigmond function)$ too it.  

The  $(Sigmond function)$ converts any number to value between (0) 1nd (1).  
The larger the input to the sigmond function, the closer the output is to (1.0) and the smaller the input, the closer the output is to (0.0).  

Since the output value is a number between (0) and (1), it can be interpreted as $Probability$.  

This is useful because we will use the logistic regression to predict the probability of something happening, instead of using a numerical value(such as price in euros or the amount of carbon emissions in kilograms).  

If we use the logistic regression in medical diagnosis, the output of the model can be for example the probability of a crtain disease.  

When there are exactly teo alternatives ('smilly face' happy or not, disease or not disease), the probability of the second outcome is of course one minus the probability of the first one.  

Two alternatives ...  
(The Probability of the second one) = 1 - (The probability of the First one).  

The model can then be used to classify different items into two classes so that we read the output from the model and iterpret it as the probability of one of the classes.  
We often label the classes 0 and 1, and let the output of the model be the probability of class 1.  

The mathimatical formula that defines the sigmond function is as follows.
$$s(z) = 1 ÷ (1 + exp(-z))$$

The function $exp$ denotes the exponential function..  
$exp(x) = e^x$, ($e$ raised to the power $x$)   
where $e$ is the so-called Euler's number that is approxiamtelly equal to (2.718). It can be computed by the Python function $exp()$ in module (math).  

### What if we have more than two classes?

You may wonder whether the logistic regression model is only applicable to two-class problems where the output can be represented as (zeros) and (ones). the good news is: it isn't. There is a simple generalisation of the sigmond function to multiple class labels, which is called softmax. We won't discuss it here but it's one more term to remeber since it is heavily used in neural networks.   

Just like in linear regression, the learning process involves adaptation of the weights or coeffecients so that the model fits the training data as well as possible.  

In the case of the standard linear regression, the optimization criterion is the squared error, and the least squares method gives us the solution in no time.  

However, in logistic regression, the outputs are probabilities and the observed reponses are binary "yes/no" labels, and the optimization criterion is something other than squared error.



### Optimization criterion in general:
The __Optimization Criterion__ is a measure or a set of conditions that a machine learning algorithm seeks to minimize ot maximize to achieve the best possible model. It defines the goal of the learning process. For example, in a typical supervised learning problem,  the criterion might be to minimize the error between the predicted and actual outputs. The specific form of the optimization criterion depends on the type of model and the nature of the problem.  

### Otimization criterion in Linear Regression(MSE):  
In __Linear Regression,__ the goal is to find line or (hyperplan in higher dimesions) that best fits data. The optimization criterion is usually the __the mean squared error (MSE)__ which measures the average square difference between the actual target values and the predicted values.  
the objective is to minimize this __(MSE),__  meaning the algorithm adjusts the model parameters (weights) to make the predictionsas close as possible to the actual values. the optimization is typically done using techniques like __gradient descent__ or __normal equation__.  

### OptimizationCriterion in Logistic Regression(Logarithmic loss):
in __Logistic Regression,__ the model predicts probabilities of binary outcomes rather than numeric values. the optimization criterion here is the __log-likelyhood__ (or more commonly, its negative, called the __log-loss__ or __cross-entropy loss__ )   

It is calculated by using the logistic regression modelto evaluate the probability of the observed label:   
If the observed label is class 1,  the probability is directly obtained from the sigmon output...  
If the observed label is class 0, the probability is obtained from one minus the sigmond output.    

#### The Logarithmic loss:
for a given set of data is the sum of negative logarithms of these probability.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Sample data: Hours studied vs. Probability of getting a raise
np.random.seed(0)
hours_studied = np.random.uniform(20, 100, 100)
raise_given = (hours_studied > 50).astype(int)  # Assume getting a raise is more likely after 50 hours of study

# Create the logistic regression model
X = hours_studied.reshape(-1, 1)
y = raise_given

# Fit the model
model = LogisticRegression()
model.fit(X, y)

# Generate a sequence of hours from min to max for prediction
X_test = np.linspace(10, 100, 300).reshape(-1, 1)
y_prob = model.predict_proba(X_test)[:, 1]

# Plot the data points and the logistic regression curve
plt.figure(figsize=(10, 6))
sns.scatterplot(x=hours_studied, y=raise_given, color='skyblue', s=100, label='Data points')
plt.plot(X_test, y_prob, color='blue', linewidth=2)

# Add labels and title
plt.xlabel('Hours studied Python')
plt.ylabel('Probability of getting a raise')
plt.ylim(-0.05, 1.05)
plt.title('Logistic Regression: Probability of getting a raise vs. Hours studied')
plt.grid(True)
plt.show()
```

Just like linear regression, logistic regression is also a simple but widely used technique.   
It has been successfully used for applications in medicine (assessing the risk of various adverse health conditions such as heart disease, cancer , aor COVID-19), social sciences (modeling factors that affect voting decisions or crime), marketing (who will click on an online ad), and countless other domains.  

the most important to realize that the most important factor in determining whether an application will be successful is often not which machine learning method you choose to use, but how you use it: the amount and the quality of the training data, the way the data is preprocessed and represented, the way the results are interpreted and applied, and so on.  

However, at some point the methods themselves may become the limiting factor. One such limit is the linearity of the model. Both linear and logistic regression are so-called linear models, as is Naive Bayes.  

An example of a problem for which linear methods are inappropriate would be predicting the chance that a user clicks an advertisment about diapers or other parenting products based on the user's age. Parents with a small babies tend to most often be somewhere between 20s and 30s. A logistic regression model where the probability of clicking the ad either increases or decreases consistantely with age is unable to output low probabilities for younger users, high probabilities for users somewhere between 20-40 years of age , and low probabilities again for older users.  
The linearity limitations can actually be curcumvented by preprocessing the data in suitable ways, by encoding the age as a "categorical" variable, which means that we include an indicator variable for each age group. This will however, lead to more parameters in the model and a higher risk of overfitting unless we have enough data.  

finding the right way to encode the data can be very tedious "manual" work. Fortunately, there is another solution. If we suspect that there may be some nonlinear patterns in the data, such as in the above diapers example, we can also apply nonlinear methods such as nearest neighbor method. Another common example is the topic of this chapter, neural networks.  

there is no clear-cut method to tell in advance which machine learning method, and which way to represent the data, will produce the best results. In practice, machine learning is a pragmatic art that involves a lot of experimentation and the best method can often only be found by trial and error
